```python
import pandas as pd
import glob # finds pathnames using pattern matching rules
```

```python
file_paths = glob.glob('/Users/priscillahandoyo/Desktop/data-science/storm-chaser/dataset/storm_data/*.csv')

dfs = []
for file in file_paths:  
    # prevent unicode decode error by trying different encodings
    for encoding in ['utf-8', 'latin-1', 'windows-1252']:
        try:
            df = pd.read_csv(file, encoding=encoding)
            dfs.append(df)
            break
        except UnicodeDecodeError:
            continue

# we want to ignore the index while we concatenating the files
combined_df = pd.concat(dfs, ignore_index=True)
# we don't want the index for the new combined file
combined_df.to_csv('stormdata_1996_2013.csv', index=False)
```
